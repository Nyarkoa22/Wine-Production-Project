# Replacing all null values with the mean of each column
---
---
### Null or missing values are a common occurrence in datasets which can pose a problem for data analysis and machine learning models. One way to handle missing values is to replace them with the mean of the corresponding feature.

## Reasons for using the mean method approach
---
<span style ="color:blue">Retains the distribution</span>: Replacing missing values with the mean ensures that the overall distribution of the data is not significantly affected. This is important because it helps to preserve the statistical properties of the dataset, which can be important for downstream analysis.

<span style ="color:blue">Simple and quick</span>: Replacing null values with the mean is a simple and quick method, especially when dealing with large datasets. This approach requires minimal computational resources, and it can be easily implemented in most programming languages and statistical software packages.

<span style ="color:blue">Avoids bias</span>: When replacing null values with the mean, we are not introducing any significant bias in the dataset. For example, if we were to replace null values with the maximum or minimum value, this could introduce a bias in the dataset, which can negatively impact the analysis and predictions made by machine learning models.

#### However, it's worth noting that replacing null values with the mean may not always be the best approach, depending on the specific characteristics of the dataset and the goals of the analysis. There are other methods available for handling missing values, such as using a regression model to predict the missing values or imputing values based on the median or mode.

```python
null_columns = ["fixed acidity", "citric acid", "chlorides",
                "free sulfur dioxide", "total sulfur dioxide",
                "density", "pH", "quality"]

for column in null_columns:
    results[column].fillna(results[column].mean(), inplace=True)
```

# Checking for Hidden Inferences using Standard scores and Outliers
---
---
### Hidden inferences are conclusions or assumptions that are not explicitly stated or obvious, but are implied or suggested by the context or language used in a statement or text

## Standard scores
### Standard scores are a type of normalized score that indicates how many standard deviations an observation is above or below the mean of a distribution. They are used to compare an individual observation to the overall distribution of scores.
### The most common type of standard score is the z-score, which is calculated by subtracting the mean of a distribution from an individual observation and dividing the result by the standard deviation of the distribution. Z_scores help us identify outliers in a dataset and understand the variability of the data

## Outliers
### Outliers are data points that are significantly different from the other observations in a dataset. They can be unusually high or low values, or values that are far away from the central tendency of the data. Outliers can occur for various reasons, such as measurement errors, data entry errors, or genuinely rare or extreme events. However, outliers can also be a result of a systematic problem in the data, such as a biased sample or a data collection method that is not appropriate for the research question.
### One common method for identifying outliers is to use statistical tests, such as the z-score test or the Grubbs' test, to detect values that are significantly different from the others (box plot).
